```python
# Define income band boundaries
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']

# Create the 'Income Band' column using pd.cut
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, include_lowest=True)
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Select specific categorical columns from the DataFrame
categorical_cols = ["Risk Weighting", "Nationality", "Occupation", "Fee Structure", 
                    "Loyalty Classification", "Properties Owned", "Income Band"]

# Dictionary to store frequency counts
frequency = {}

# Loop through each categorical column
for col in categorical_cols:
    frequency[col] = df[col].value_counts()  # store value counts
    print(f"\nValue Counts for '{col}':")
    display(frequency[col])  # display value counts
```

```python
frequency["Risk Weighting"].plot(kind='bar')
```

```python
df["GenderId"].value_counts()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(10,5))
    sns.countplot(data=df,x=col,hue='GenderId')
    plt.title(f"Distribution of '{col}' by GenderId")
    plt.xticks(rotation=45)
    plt.show()
```

```python
categorical_cols2 = ["Risk Weighting",  "Occupation", "Fee Structure","Loyalty Classification", "Properties Owned", "Income Band"]
for i, col in enumerate(categorical_cols2):
    plt.figure(figsize=(10,5))
    sns.countplot(data=df,x=col,hue='Nationality')
    plt.title(f"Distribution of '{col}' by Nationality")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Numerical col 
numerical_cols = ['Fee Structure','Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[col], kde=True) # To visualize the distribution of numerical data.
    plt.title(col)                  # It shows how frequently values occur by binning them into intervals.
plt.tight_layout()                  # KDE helps understand the underlying distribution shape 
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)  
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
